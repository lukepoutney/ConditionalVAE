# Resize Graph

```python
import csv
import matplotlib
import matplotlib.pyplot as plt
```

```python
name = 'SegLargeQty0'
with open('outputs/'+name+'/loss'+name+'.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    loss = list(csv_reader)
    loss = loss[1:]
    train_loss = [float(item[0]) for item in loss]
    val_loss = [float(item[1]) for item in loss]

print(train_loss)
print(len(train_loss))
print(val_loss)
print(len(val_loss))
epochs = 100
```

```python
plt.figure(figsize=(10,10))
plt.plot(range(1,len(train_loss)+1), train_loss, label="Train Loss")
plt.plot(range(1,len(train_loss)+1), val_loss, label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.ylim([11300,11400])
axes = plt.gca()
plt.savefig('outputs/'+name+'/loss'+name+'resize.png')
```
