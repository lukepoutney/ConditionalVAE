```python
import csv
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams.update({'font.size': 14})
```

```python
name = 'DualReconstructionProceduralImages0'
with open('outputs/'+name+'/loss'+name+'.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    loss = list(csv_reader)
    loss = loss[1:]
    train_loss = [float(item[0]) for item in loss]
    val_loss_true = [float(item[1]) for item in loss]
    val_loss_noisy = [float(item[2]) for item in loss]

# print(train_loss)
# print(len(train_loss))
# print(val_loss)
# print(len(val_loss))
epochs = len(train_loss)
```

```python
plt.figure(figsize=(10,10))
plt.plot(range(1,len(train_loss)+1), train_loss, label="Train Loss")
plt.plot(range(1,len(train_loss)+1), val_loss_true, label="Validation Loss (True Reconstruction)")
plt.plot(range(1,len(train_loss)+1), val_loss_noisy, label="Validation Loss (Noisy Reconstruction)")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.title("Conditional VAE with Dual Conditons")
plt.ylim([0,500])
axes = plt.gca()
plt.savefig('outputs/'+name+'/loss'+name+'resize.png')
```
